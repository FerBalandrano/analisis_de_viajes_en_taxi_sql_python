## Pruebas de hipótesis

- Hipotesis nula: No hay diferencia en la duración promedio de viajes entre días con condiciones clímaticas buenas y días con condiciones clímaticas malas, en sábados.
- Hipotesis alternativa: Sí hay diferencia en la duración promedio de viajes entre días con condiciones clímaticas buenas y días con condiciones clímaticas malas, en sábados.

```python
# Creación de una columna en el df de los viajes desde Loop al Aeropuerto de O'Hare para poder comparar sábados con sábados.
loop_ohare_df['day_of_week'] = loop_ohare_df['start_ts'].dt.day_name()
```

```python
# Verificación de cuáles DOW tenemos en el dataset.
print("Días disponibles:")
print(loop_ohare_df['day_of_week'].value_counts())
```

```python
# Separación de los datos por condición climática.
good_weather = loop_ohare_df[loop_ohare_df['weather_conditions'] == 'Good']['duration_seconds']
bad_weather = loop_ohare_df[loop_ohare_df['weather_conditions'] == 'Bad']['duration_seconds']

print(f"Viajes con buen clima: {len(good_weather)}")
print(f"Viajes con mal clima: {len(bad_weather)}")
```

```python
# Obtención de estadísticas generales de ambos grupos.
good_weather.describe()
```

```python
bad_weather.describe()
```

### Observaciones
- Bad weather tiene una mayor duracion promedio de viaje, pero menor máximo y menor minimo, lo que indica menor variabilidad entre los viajes en bad weather.
- El cuartil 50% de bad está cerca del cuartil 75% de good, lo que nos habla de que en general los viajes en bad weather tienden a durar más que los viajes en good weather.

```python
# Creación de histogramas para comparar distribuciones de ambos grupos.
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(good_weather, bins=30, alpha=0.7, label='Good Weather')
plt.title('Distribución - Buen Clima')
plt.xlabel('Duración (segundos)')

plt.subplot(1, 2, 2)
plt.hist(bad_weather, bins=30, alpha=0.7, label='Bad Weather', color='orange')
plt.title('Distribución - Mal Clima')
plt.xlabel('Duración (segundos)')

plt.tight_layout()
plt.show()
```

```python
# Elijo la prueba U de Mann-Whitney ya que la distribución de mis datos no parece del todo normal (no es paramétrica) y funciona mejor para muestras que que no son normales (o de las que no podemos asumir que son normales).
alpha = 0.05 

statistic, p_value = mannwhitneyu(bad_weather, good_weather, alternative='two-sided')

print(f"Estadístico U: {statistic}")
print(f"P-valor: {p_value}")
print(f"Nivel de significación: {alpha}")

# Interpretación.
if p_value < alpha:
    print("Rechazamos H₀: Hay diferencia significativa entre la duración de los viajes en días con buen clima vs días con mal clima.")
else:
    print("No rechazamos H₀: No hay evidencia de diferencia significativa entre la duración de los viajes en días con buen clima vs días con mal clima.")
```

Muy bien con el uso de un test Mann-Whitney U, sin embargo, debes tener en cuenta que ese test realiza una comparación de las distribuciones, no de las medias, por lo que tampoco es el más adecuado para lo solicitado.

Si bien las variables a testear no distribuyen normal, también se puede aplicar un t-test. Para este test, en la literatura comúnmente se menciona que se requiere normalidad de la variable, sin embargo, lo que en realidad se necesita es la normalidad de la media muestral (si la variable distribuye normal, eso se cumple directamente). Para ello, basta con tener una muestra con "n grande" para que, según el teorema central del límite, la distribución de la media muestral sea aproximadamente normal y con ello se satisface el requerimiento para el t-test.
